```python
from matplotlib import pyplot as plt
```

```python
x = [2,5,1,2]
y = [1,4,3,1]
```

```python
plt.plot(x, y, c="red", linestyle="--",linewidth=2, marker="o", alpha=0.5)
plt.plot(y, x, c="#30244f", linestyle = ":", linewidth=2, marker=".",markersize=20, alpha= 1)

plt.xlim(-3,9)
plt.ylim(-3,9)
plt.plot()
```
