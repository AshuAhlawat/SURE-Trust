```python
import pandas as pd
from matplotlib import pyplot as plt 
import numpy as np
import torch
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv("./life_expectancy.csv")
#dataset=dataset.dropna(axis=0)

dataset=dataset.drop(["Country","Year"],axis=1).dropna()[:500]
print(dataset.columns)
label=dataset["Life"]
label1=np.array(label).reshape(-1,1)
labels=torch.tensor(label1,dtype=torch.float32)
fea=dataset.drop("Life",axis=1)
fea1=pd.get_dummies(fea)
fea2=np.array(fea1)
fea3= StandardScaler().fit_transform(fea2)
features=torch.tensor(fea3,dtype=torch.float32)
print(features)
print(labels)
```

```python
model=torch.nn.Sequential(
torch.nn.Linear(20,10),
torch.nn.ReLU(),
torch.nn.Linear(10,1)
)

for i in model.parameters():
    print(i)
```

```python
preds=model(features)
print(preds[:5])
```

```python
from torch.nn.functional import mse_loss
loss=mse_loss(preds,labels)
print(loss)
```

```python
loss.backward()
```

```python
optimizer=torch.optim.SGD(model.parameters(),lr=1e-12)
optimizer.step()
preds=model(features)
loss=mse_loss(preds,labels)
print(loss)
```

```python
optimizer=torch.optim.SGD(model.parameters(),lr=1e-3)
epochs=10000
for epoch in range(epochs):
    preds=model(features)
    loss=mse_loss(preds,labels)
    if(epoch+1)%1000==0:
        print(loss)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
```
