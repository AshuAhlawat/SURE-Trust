```python
import pandas as pd
import numpy as np

data = pd.read_csv("../weather.csv")[:50]

print(data.columns)
features = np.array(data["Temperature (C)"])
labels = np.array(data["Humidity"])
# labels = 1/labels
```

```python
from matplotlib import pyplot as plt

plt.scatter(features,labels)
plt.show()
```

```python
from numpy import random
weight = random.randn()
bias = random.randn()

print(weight, bias)
```

```python
import numpy as np
def mse(predictions ,labels = labels):
    diff = labels-predictions 
    loss = (np.sum(diff**2))/len(labels)
    return loss
# mse(np.array([1,2,3,4]),np.array([2,3,2,5]))
```

```python
predictions = weight*features + bias
print(predictions[:5])
```

```python
loss = mse(predictions, labels)
print(loss)
```

```python
plt.scatter(features,labels)
plt.plot(features,predictions, color = "red")
plt.show()
```

```python
weight_grad = np.sum(-1*features*(labels- predictions))/len(features)
bias_grad = np.sum(-1*(labels- predictions))/len(features)

print(weight, bias)
print(weight_grad, bias_grad)
```

```python
lr = 0.01

if weight_grad < 0:
    weight += lr
else:
    weight -= lr
if bias_grad < 0:
    bias += lr
else:
    bias -= lr
```

```python
predictions = weight*features + bias
loss = mse(predictions, labels)
print(loss)
```

```python
plt.scatter(features,labels)
plt.plot(features,predictions, color = "red")
plt.show()
```
