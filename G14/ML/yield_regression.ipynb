```python
import numpy as np
```

```python
# features Temp, Rainfall, Humidity
features = np.array([
    [73, 67, 43],
    [91, 88, 64],
    [87, 134, 58],
    [102, 43, 37],
    [69, 96, 70]
])

# labes Apples , Oranges
labels = np.array([
    [56, 70],
    [81, 101],
    [119, 133],
    [22, 37],
    [103, 119]
])

label_1 = labels[:,0]
label_2 = labels[:,1]
print(label_1,label_2)
```

```python
weights = np.random.randn(3)
bias = np.random.randn()

print(weights, bias)
```

```python
def predict(weights, bias):
    x = features*weights
    
    preds = np.zeros(len(features))
    for i in range(len(x)):
        preds[i] = np.sum(x[i])

    preds += bias

    return preds

preds = predict(weights, bias)

print(preds)
```

```python
def mse(preds, labels):
    return np.sum((preds - labels)**2)/len(labels) # mean squared error
loss = mse(preds, label_1)
print(loss)

# loss.backward()
# w3.grad
# w1.grad
# w2.grad
```

```python

diff = label_1 - preds

x1 = features[:, 0]
x2 = features[:, 1]
x3 = features[:, 2]

w1_grad = np.sum(-1 * diff * x1)/len(x1)
w2_grad = np.sum(-1 * diff * x2)/len(x1)
w3_grad = np.sum(-1 * diff * x3)/len(x1)

b_grad = np.sum(-1 * diff)/len(x1)

w_grad = np.array([w1_grad, w2_grad, w3_grad])
print(w_grad, b_grad)
```

```python
lr = 0.00001
weights = weights + (-1 * w_grad * lr)
bias = bias + (-1 * b_grad * lr)

print(weights, bias)
```

```python
preds = predict(weights,bias)
loss = mse(preds, label_1)

print(loss)
```

```python
lr = 0.00001

for i in range(1000):
    diff = label_1 - preds

    x1 = features[:, 0]
    x2 = features[:, 1]
    x3 = features[:, 2]

    w1_grad = np.sum(-1 * diff * x1)/len(x1)
    w2_grad = np.sum(-1 * diff * x2)/len(x1)
    w3_grad = np.sum(-1 * diff * x3)/len(x1)

    b_grad = np.sum(-1 * diff)/len(x1)

    w_grad = np.array([w1_grad, w2_grad, w3_grad])
    # print(w_grad, b_grad)

    weights = weights + (-1 * w_grad * lr)
    bias = bias + (-1 * b_grad * lr)

    preds = predict(weights,bias)
    loss = mse(preds, label_1)

    print(loss)
```
