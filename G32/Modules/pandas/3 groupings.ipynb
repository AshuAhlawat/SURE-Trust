```python
import pandas as pd
```

```python
port_df = pd.read_csv("./Port_Data.csv", index_col="Index")

# port_df.head()
port_df.sort_values(by=["Expected Arrivals"], ascending=False, ignore_index=True, inplace= True)

# port_df["Country"].str.upper().value_counts()
port_df["Country"].unique()[0]
```

```python
# print out the top 5 countries with the highest gap between the departures and arrivals
```

```python
for country , c_df in port_df.groupby("Country"):
    total_arrivals = sum(c_df["Expected Arrivals"])
    
    print(f"{country} : {total_arrivals}")
    

c_df
```
