```python
import matplotlib.pyplot as plt
```

```python
x = {
    "S. Engineer" : 10,
    "Marketing" : 30,
    "Testing" : 35,
    "Data Analyst" : 25,
    "CTO" : 70,
    "Tech Lead" : 55
}
```

```python
plt.pie(x.values(), labels=x.keys(),startangle=180,colors=("Red","Blue"), explode=(0,0,0.3,0,0,0), autopct="%0.1f%%")
plt.show()
```
