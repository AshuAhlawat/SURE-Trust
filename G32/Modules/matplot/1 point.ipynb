```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,-1,1]


plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(y,x, s=[20,80,10], c= ["Red", "Blue", "Black"], marker="x", alpha=0.5)
plt.scatter(x,y)
plt.ylabel("Height", c = "Red", size = 20)
plt.xlabel("Length", c = "Green")

plt.ylim(0,5)
plt.xlim(-3,5)

plt.title("Plot", c="Pink", size = 50)

plt.show()
```
